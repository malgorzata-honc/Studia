```python
from sklearn.datasets import load_iris
data = load_iris()
X = data.data
Y = data.target
```

```python
import matplotlib.pyplot as plt
```

ZADANIE 1

```python
all_pairs = [(i, j) for i in range(X.shape[1]) for j in range(X.shape[1]) if i < j]
all_pairs
```

```python
for pair in all_pairs:
    feat1, feat2 = pair
    
    plt.figure(figsize=(9, 3.5))

    plt.subplot(121)
    plt.plot(X[Y==0, feat1], X[Y==0, feat2], "yo", label="Iris setosa")
    plt.plot(X[Y==1, feat1], X[Y==1, feat2], "bs", label="Iris versicolor")
    plt.plot(X[Y==2, feat1], X[Y==2, feat2], "g^", label="Iris virginica")
    plt.xlabel(data.feature_names[feat1], fontsize=14)
    plt.ylabel(data.feature_names[feat2], fontsize=14)
    plt.legend(fontsize=12)

    plt.subplot(122)
    plt.scatter(X[:, 2], X[:, 3], c="k", marker=".")
    plt.xlabel(data.feature_names[feat1], fontsize=14)
    plt.tick_params(labelleft=False)

    #save_fig("classification_vs_clustering_plot")
    plt.show()
```

ZADANIE 3

```python
from sklearn.cluster import KMeans
from scipy import stats
import numpy as np
```

```python
all_pred = {}

for pair in all_pairs:
    feat1, feat2 = pair
    k = 3
    kmeans = KMeans(n_clusters=k, random_state=42)
    y_pred = kmeans.fit_predict(X[:,[feat1,feat2]])
    
    mapping = {}
    for class_id in np.unique(Y):
        mode, _ = stats.mode(y_pred[Y==class_id], keepdims=True)
        mapping[mode[0]] = class_id
    
    y_pred = np.array([mapping[cluster_id] for cluster_id in y_pred])
    all_pred[(feat1, feat2)] = {'pred': y_pred, 'inertia': kmeans.inertia_}

all_pred
```

ZADANIE 4

```python
all_diff = {}
for feat1, feat2 in all_pairs:
    y_pred = all_pred[(feat1, feat2)]['pred']
    diff = np.sum(Y != y_pred)
    all_diff[(feat1, feat2)] = diff
    print(f"Dla cech {data.feature_names[feat1]} i {data.feature_names[feat2]} iloÅ›Ä‡ bÅ‚Ä™dnie pogrupowanych: {diff}")
```

```python
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
y_pred = kmeans.fit_predict(X)
    
mapping = {}
for class_id in np.unique(Y):
    mode, _ = stats.mode(y_pred[Y==class_id], keepdims=True)
    mapping[mode[0]] = class_id
    
y_pred = np.array([mapping[cluster_id] for cluster_id in y_pred])
all_feat_diff = np.sum(y_pred != Y)
all_feat_diff
```

ZADANIE 9

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=35, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=2, linewidths=12,
                color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

```python
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X[:,[2,3]])
plt.figure(figsize=(8, 4))
plot_decision_boundaries(kmeans, X[:,[2,3]])
plt.show()
```

```python
import math
max_inertia = (None, 0)
min_inertia = (None, math.inf)
max_diff = (None, 0)
min_diff = (None, math.inf)

for features, inertia in all_inertia:
    if inertia > max_inertia[1]:
        max_inertia = (features, inertia)
    if inertia < min_inertia[1]:
        min_inertia = (features, inertia)

for features, diff in all_diff.items():
    if diff > max_diff[1]:
        max_diff = (features, diff)
    if diff < min_diff[1]:
        min_diff = (features, diff)
```

ZADANIE 14

```python
x = X[:,[2,3]]

kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(x)
                for k in range(1, 10)]
inertias = [model.inertia_ for model in kmeans_per_k]
```

```python
plt.figure(figsize=(8, 3.5))
plt.plot(range(1, 10), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)
plt.annotate('Elbow',
             xy=(2, 100),
             xytext=(0.55, 0.55),
             textcoords='figure fraction',
             fontsize=16,
             arrowprops=dict(facecolor='black', shrink=0.1)
            )
plt.axis([1, 8.5, 0, 1300])
plt.show()
```

ZADANIE 15

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = [silhouette_score(x, model.labels_)
                     for model in kmeans_per_k[1:]]
```

```python
plt.figure(figsize=(8, 3))
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.axis([1.8, 8.5, 0.55, 0.85])
plt.show()
```

ZADANIE 16

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter
import matplotlib as mpl

plt.figure(figsize=(11, 9))

for k in (2, 3):
    plt.subplot(2, 2, k -1)
    
    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(x, y_pred)

    padding = len(x) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (2, 3):
        plt.ylabel("Cluster")
    
    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)

plt.show()
```
